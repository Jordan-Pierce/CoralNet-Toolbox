```python
data_dir = "C:\\Users\\jordan\\Documents\\GitHub\\CoralNet-Toolbox\\data\\mir\\sites\\CHCA_CC\\datasets\\semantic_tiled_2048"
```

```python
from ultralytics import YOLO
import torch
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
# device = "cpu"  # Force CPU for debugging
print(f"Using device: {device}")
```

```python
data_yaml = f"{data_dir}\\data.yaml"
results_dir = f"{data_dir}\\results"

name = "yolo26m"
```

```python
model = YOLO("yolo26m-semseg.yaml")
model.task
```

```python
model.train(data=data_yaml, epochs=300, imgsz=1024, project=results_dir, name=name, exist_ok=True, device=device)
```

```python
split = 'val'

model.val(data=data_yaml, imgsz=1024, device=device, split=split, project=f"{results_dir}/{name}", name=split, exist_ok=True)
```

```python
split = 'test'

model.val(data=data_yaml, imgsz=1024, device=device, split=split, project=f"{results_dir}/{name}", name=split, exist_ok=True)
```

```python
import glob 

files = glob.glob(f"{data_dir}\\test\\images\\*.*")
print(len(files))
```
