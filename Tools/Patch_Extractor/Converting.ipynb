```python
from patch_extractor_utils import *

import glob
import random
import matplotlib.pyplot as plt
```

```python
# Folder containing all the txt files output by Patch Extractor tool
folder = "TXT_Files"

# Patch Extractor tool provides the xy location of top-left corner
# To make the points centered in the patch, patch size needs to be known.
patch_size = 224
```

```python
files = glob.glob(folder + "/*.txt")
df = pd.concat([txt_to_csv(file, patch_size=224) for file in files])
```

```python
df.sample(10)
```

```python
sample = df[df['Name'] == df.iloc[random.randint(0, len(df))]['Name']]

image_path = sample.iloc[0]['File_Path']
y = sample['Row'].values
x = sample['Column'].values

plt.figure(figsize=(10, 10))
plt.scatter(x, y, c='red')
plt.imshow(plt.imread(image_path))
plt.show()
```
